```python
import pandas as pd
import datetime as dt
from src.tasks import BondsDescription
```

```python
pos_df = pd.concat( pd.read_excel(fr'C:\Users\SSamoliuk\Downloads\Bond PnL Report_FI_{entity}_2026-02-27.xlsx', skiprows=1).assign(Entity=entity) for entity in ['NRBM', 'CAAC'] )
pos_df['vol_bio'] = (pos_df['Quantity'] * 1e-6).round(2)
```

```python
desc_df = BondsDescription(date=dt.date.today()).read_output()
```
